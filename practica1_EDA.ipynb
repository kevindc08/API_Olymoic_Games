```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
summer = pd.read_csv('Data/Athletes_summer_games.csv')
winter= pd.read_csv('Data/Athletes_winter_games.csv')
regions = pd.read_csv('Data/regions.csv')
df = pd.concat([summer, winter])
del summer, winter
```

```python
variables=list(diccionario_nulos.keys())
nulos=list(diccionario_nulos.values())
plt.bar(variables,nulos)
plt.xlabel('Variable')
plt.ylabel('Total de nulos')
plt.title('Nulos por columna')
plt.xticks(rotation=90)
plt.show()
```

```python

Data=df["Name"].value_counts().head(10)
plt.bar(Data.keys(),Data.values)
plt.xlabel('Atletas')
plt.ylabel('Participacion')
plt.title('10 Atltas con mas participaciones en juegos olimpicos')
plt.xticks(rotation=90)
plt.show()
```

```python
Data=df["Sex"].value_counts()
plt.pie(labels=Data.keys(),x=Data.values, autopct='%1.1f%%')
plt.title('Porcentaje de sexos que articiparon')
plt.xticks(rotation=90)
plt.show()
```

```python
df['Age_R']=pd.cut(x=df['Age'],bins=np.arange(10,110,10))
Data = df['Age_R'].value_counts()

keys=[interval.left for interval in Data.index]
values=Data.values
plt.bar(keys,values)
plt.yscale('log')
plt.xlabel('Edad de atletas')
plt.ylabel('numero de atletas(escala logaritmica)')
plt.title('Edades de los atletas que participan en juegos olimpicos')
plt.show()
```
